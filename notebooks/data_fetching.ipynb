```python
import requests

url = "https://cricbuzz-cricket.p.rapidapi.com/mcenter/v1/108811/scard"

headers = {
	"x-rapidapi-key": "8f3de2cf84msh4e512ffc04f9defp18f43bjsn78b7858eac5b",
	"x-rapidapi-host": "cricbuzz-cricket.p.rapidapi.com"
}

response = requests.get(url, headers=headers)

print(response.json())
data = response.json()
```

```python
for scorecard in data.get("scorecard", []):
        battting = []
        for batsman in scorecard.get("batsman",[]):
            battting.append({
                "Batsman": batsman["name"],
                "Runs": batsman["runs"],
                "Balls": batsman["balls"],
                "4s": batsman["fours"],
                "6s": batsman["sixes"],
                "SR": batsman["strkrate"]
            })
df = pd.DataFrame(battting)
df.set_index(df["Batsman"], inplace=True)
df
```

```python
for innings  in data.get("scorecard", []):         
    fallofwickets = []
    fows = innings.get("fow", {})
    if not fows:
        continue
    for fow in fows.get("fow", []):
        fallofwickets.append({
            "name": fow["batsmanname"],
            "score": fow["runs"],
            "over": fow["overnbr"]
        })
    df = pd.DataFrame(fallofwickets)
    print(df)
```

```python
def apicall(url):
    try:
        response = requests.get(url, headers=headers)
        return response.json()
    except Exception as e:
        return e


venuelist = []

for i in range(1, 2):
    url = f"https://cricbuzz-cricket.p.rapidapi.com/venues/v1/{i}"
    data = apicall(url)

    # skip invalid responses
    if not data:
        continue

    ground = data.get("ground")
    city = data.get("city")
    country = data.get("country")
    capacity = data.get("capacity")
    homeTeam = data.get("homeTeam")

    # skip empty venues
    if not ground:
        continue

    venuelist.append((ground, city, country, capacity, homeTeam))
```
